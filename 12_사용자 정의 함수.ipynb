# 1. 함수
- 변수 x, y에 대해 x값이 정해지면 그에 따라 y값을 결정하는 것

### 1-1. 이름만 있는 함수
```
def 함수명():
  함수가 호출되면 실행할 문장
  ...
```
```
함수명() # 호출
```

````python
def func1():
  print('처음으로 만드는 함수!')
````

````python
func1()
func1()
````

````python
for i in range(5):
  func1()
````

### 1-2. 매개변수(parameter)가 있는 함수
```
def 함수명(변수1, 변수2, ...):
  함수가 호출되면 실행할 문장
  ...
```
```
함수명(값1, 값2 ...) # 호출
```

````python
def func2(n):
  print('입력받은 숫자: %d' % n)
````

````python
num1 = int(input('숫자를 입력해보아요!'))

func2(num1)
func2(5)
````

````python
def func3(start, end):
  sum = 0
  for i in range(start, end+1):
    sum += i
  print('%d부터 %d까지의 합: %d' % (start, end, sum))
````

````python
func3(1, 100)
````

```
```
변수 = 함수명(값1, 값2 ...) # 호출

````python
def func4():
  return '다음 주 월요일은 휴일!'
````

````python
result = func6(10, 5)
print(result)
````

### 1-4. 디폴트 매개변수

```
def 함수명(변수1=값1, 변수2 = 값2, ...)
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(값1, 값2 ...) # 호출
변수 = 함수명() # 호출
```

````python
def func7(num1, num2):
  return num1 * num2
````

````python
print(func7(10, 3))
````

````python
def func8(num1=5, num2=3):
  return num1 * num2
````

````python
print(func8(10, 3))
````

````python
print(func8())
````

### 1-5. 매개변수의 개수가 가변적인 함수

```
def 함수명(*변수):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(값1, 값2 ...) # 호출
변수 = 함수명(값1) # 호출
```
> 매개변수는 함수 안에서 튜플형태로 사용

````python
def func9(*args):
  result = 0
  for i in args:
    result += i
  return result
````

````python
print(func9(10))
print(func9(10, 30, 20, 70, 50))
````

### 1-6. 매개변수를 딕셔너리 구조로 전달받는 함수

```
def 함수명(**변수):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(변수1=값1, 변수2=값2 ...) # 호출
```

````python
def func10(**args):
  for i in args:
    if i == 'userid':
      id = args[i]
  return id
````

````python
user = func10(userid='hogeun4656', name='류호근', age=20)
print(user)
````

### 1-7. 리턴값이 여러개인 함수

```
def 함수명(변수1, 변수2 ...):
    함수가 호출되면 실행할 문장
    ...
    return 값1, 값2, ...
```
```
변수1, 변수2, ... = 함수명(변수1, 변수2 ...) # 호출
```

````python
def func11(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
result1, result2, result3, result4 = func11(10, 3)
print(result1, result2, result3, result4)
````

````python
result1 = func11(10, 3)
print(result1[2])
````

````python
_, _, result1, _= func11(10, 3) # 곱셈의 결과만 필요할 때
print(result1)
````
