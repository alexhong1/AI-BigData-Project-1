# 1. 문자열 인덱싱


- 문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능

```python
str = '문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능'
print(str)
```

```python
print(str[0])
print(str[1])
print(str[4])
print(str[-1]) # 마지막 문자
print(str[-2])

# 저장공간과 값의 차이 한 번 더 생각해보기! 
print('류호근짱'[3])
```

# 2. 문자열 슬라이싱

- 문자열에서 원하는 부분을 추출하는 기능

```python
str = '문자열에서 원하는 부분을 추출하는 기능'
print(str)
```

```python
print(str[0:4]) # 인덱스 0부터 4 이전까지 문자열
print(str[6:13])
print(str[:13]) # 처음부터 13 직전까지
print(str[6:]) # 6 부터 끝까지
```

# 3. 문자열 함수

- 문자열과 관련된 여러가지 기능(함수)

```python
str = '문자열 함수는 문자열과 관련된 여러가지 기능'
```

```python
# count() : 문자열에서 매개변수로 전달된 문자열의 출현 횟수를 반환
print(str.count('문'))
print(str.count('관련된'))
print(str.count('호근짱'))
```

```python
# find() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환
print(str.find('문'))
print(str.find('웤'))

# print(str.find('문자')) # 쓰는 목적이 다르다.
```

```python
# index() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환
print(str.index('자'))
print(str.index('꿕')) # 없으면 에러 ValueError : substring not found
```

```python
# replace() : 특정 문자열을 치환
print(str)
print(str.replace('기능', '함수'))
print(str.replace('기능', '함수').count('함수'))
```

```python
# upper() : 문자열을 모두 대문자로 변환
# lower() : 문자열을 모두 소문자로 변환
str = '문자열 function는 문자열과 관련된 여러가지 FUNCTION'

print(str.upper())
print(str.lower())
```

```python
# lstrip() : 문자열의 왼쪽 공백을 제거
# rstrip() : 문자열의 오른쪽 공백을 제거
# strip() : 문자열의 양쪽 공백을 제거
str = '             hello              python                  '
print(str)
print(str.lstrip())
print(str.rstrip())
print(str.strip())
print(str.replace(' ', ''))
```

```python
# join() : 문자열에 특정 문자열을 삽입
str = '너의모든순간'
print(' '.join(str))
```
