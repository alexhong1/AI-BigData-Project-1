# 1. scope

- 변수의 접근할 수 있는 범위
- local: 가장 가까운 함수 안 범위
- global : 함수 바깥의 변수 또는 import된 module

```python
num1 = 10

def func1():
  num2 = 20
  print(num2)
```

```python
print(num1)
print(num2) # NameError: name 'num2' is not defined
```

```python
# locals() : 로컬 변수를 확인하는 함수
# globals() : 글로벌 변수를 확인하는 함수

num1 = 10

def func1():
  num2 = 20
  print('num1 로컬 변수: ', 'num1' in locals())
```

```python
func1()
```

```python
num1 = 10

def func1():
  num2 = 20
  print('num1 로컬 변수: ', 'num1' in globals())
```

```python
func1()
```

```python
num1 = 10

def func2():
  num1 = 20 # 로컬변수에 값 저장
  print(num1)
```

```python
func2()
print(num1)
```

# 2. global 키워드

- 함수 내부에서 로컬변수가 아닌 글로벌 변수로 사용하게 함

```python
num1 = 10

def func3():
  print(num1)

def func4(num):
  num1 = num # num1 이 로컬 변수로 쓰임
```

```python
func3()
func4(5)
func3()
```

```python
num1 = 10

def func3():
  print(num1)

def func4(num):
  global num1
  num1 = num
```

```python
func3()
func4(5)
func3()
```
