# 2. 조건문

### 2-1. if문

````python
# 비교(관계) 연산자
# <, >, <=, >=, ==, !=
print(10 > 5)
print(10 < 5)
print(10 >= 5)
print(10 <= 5)
print(10 == 5) # 같니?
print(10 != 5) # 다르니?
````

```
if 조건식(결과가 True 또는 False가 나오는 문장):
    조건식이 True인 경우 실행할 문장
```

````python
age = int(input('나이를 입력하세요 : '))

# 들여쓰기 신경쓰기!
if age > 19 :
  print('술 가져와')
  print('나! 성인이야!')

print('입력한 나이 : ', age)
````

````python
age = int(input('나이를 입력하세요 : '))

if age > 19 : print('술 가져와')

print('입력한 나이 : ', age)
````

### 2-2. if ~ else 문

```
if 조건식(결과가 True 또는 False가 나오는 문장) :
    조건식이 True인 경우 실행할 문장
else :
    조건식이 False인 경우 실행할 문장
```

````python
age = int(input('나이를 입력하세요 : '))

# 들여쓰기 신경쓰기!
if age > 19 :
  print('술 가져와')
  print('나! 성인이야!')
else :
  print('나가')

print('입력한 나이 : ', age)
````

### 2-3. if ~ elif ~ else()문

```
if 조건식1(결과가 True 또는 False 가 나오는 문장):
    조건식1이 True인 경우 실행할 문장
elif 조건식2 :
    조건식2가 True인 경우 실행할 문장
elif 조건식3 :
    조건식3이 True인 경우 실행할 문장
...
else :
    조건식이 모두 False인 경우 실행할 문장
```

````python
age = int(input('나이를 입력해보아요!'))

if age > 19:
  print('성인입니다!')
elif age > 14:
  print('청소년입니다!')
elif age > 6:
  print('어린이입니다')
elif age > 0:
  print('유아입니다')
else:
  print('제대로 나이를 입력한 거 맞나요...?')  
print('입력한 나이:', age)
````

### 문제
- 국어점수, 영어점수, 수학점수를 입력 받아 총점, 평균, 학점을 구하는 프로그램을 작성해보자.
(단, 학점은 평균 기준이며 90점 이상은 A학점, 80점 이상은 B학점, 70점 이상은 C학점, 60점 이상은 D학점, 나머지는 F학점으로 한다.)

````python
kor, eng, math = map(int, input('국어, 영어, 수학점수를 입력하세요.').split(' '))

total = kor + eng + math
avg = total / 3
# 학점을 담을 변수
grade = ''

if avg >= 90:
  grade = 'A'
  print('A학점이지롱')
elif avg >= 80:
  grade = 'B'
  print('B학점이지롱')
elif avg >= 70:
  grade = 'C'
  print('C학점이지롱')
elif avg >= 60:
  grade = 'D'
  print('D학점이지롱')
else:
  grade = 'F'
  print('F학점이지롱')

print(f'총점 : {grade}, 평균 : {avg:.2f}')
print('총점 : ', total)
print('평균 : ', avg)
print('학점 : ', grade)
````

### 문제 2.
- 숫자를 입력 받아 2의 배수인지, 3의 배수인지 아니면 2, 3의 배수가 모두 아닌지를 판별하는 프로그램을 작성해보자

````python
num = int(input('숫자를 입력!'))

if num % 6 == 0:
  print(f'{num}은 2와 3의 배수')
elif num % 2 == 0:
  print(f'{num}은 2의 배수')  
elif num % 3 == 0:
  print(f'{num}은 3의 배수')
else :
  print(f'{num}은 2, 3 배수 둘 다 아니지롱~!')
````

````python
# 삼항연산자 문법을 활용해서 한 줄에 다 적을 수는 있으나
# 코드의 가독성이 떨어질 수도 있기 때문에 유의!
num = int(input('숫자를 입력하세요 : '))

print(f'{num}은 2의 배수') if num % 2 == 0 else print(f'{num}은 3의 배수') if num % 3 == 0 else print(f'{num}은 2, 3 배수 둘 다 아니지롱~!')
````

### 2-5. 논리 연산자(AND)
- 양쪽 두 조건식의 값이 모두 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|


````python
# 로그인 프로그램 만들기
# 아이디 : admin, 비밀번호 : 1234
userid = input('아이디를 입력해보아요!')
userpw = input('비밀번호 입력해보아요!')

if userid == 'admin' and userpw == '1234':
  print('로그인 되었습니다!')
else:
  print('아이디 또는 비밀번호를 확인하세요!')
````

### 2-6. 논리 연산자(OR)
- 양 쪽 두 식의 값이 하나라도 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|

````python
# 주민등록번호에 의한 성별 판별
serialNo = input('주민등록번호 13자리를 -없이 입력하세요!')

if serialNo[6] == '1' or serialNo[6] == '3' or serialNo[6] == '5':
  print('남자에요!')
else:
  print('여자에요!')
````
