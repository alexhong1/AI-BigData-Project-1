### 1-1. 클래스와 객체
- 건축 설계도가 클래스라면, 실제로 지어진 집은 객체로 비유
- 객체는 클래스로 생성되어 구체화된 인스턴스
- 실제로 클래스가 인스턴스화 되어 메모리에 상주하는 상태를 객체라고 부름
- 파이썬의 모든 변수는 객체를 저장

### 1-2. 클래스 만들기

```
class 클래스명 : 
  def __init__(self):
      self.필드명1 = 값1
      self.필드명2 = 값2
      ...
      객체가 메모리에 로드될 때 가장 먼저 실행될 문장
      ...
  def 메소드명(변수1, 변수2 ...):
      메소드가 호출되면 실행될 문장
      ...
```
- 생성자는 클래스를 객체화 시킬 때 가장 먼저 실행되는 함수
- 클래스를 통해 호출되는 변수를 필드라고 함
- 클래스를 통해 호출되는 함수를 메소드라고 함

````python
# 기능이 없는 클래스
class Dog:
  pass # 내용이 없는 블록을 만들 때 사용
````

### 1-3. 생성자
- 클래스 인스턴스가 생성될 때 자동으로 호출됨
- __ init __(self, 변수1, 변수2 ..):
- self 매개변수는 항상 첫번째 오며, 자기 자신의 객체를 가리킴
- 이름이 꼭 self일 필요는 없지만 관례적으로 self를 사용
- 생성자에서는 해당 클래스가 다루는 데이터를 정의

````python
class Dog:
  def __init__(self):
    print(self, 'init 호출')
````

````python
Jin = Dog()
print(Jin)

Jin2 = Dog()
print(Jin2)
````

````python
class Dog:
  def __init__(self):
    print(self, 'init 호출')
    self.name = '이름없음'
    self.age = 0
````

````python
Jin = Dog()
print(Jin)
print(Jin.name)
print(Jin.age)
````

````python
Jin.name = '진순이'
Jin.age = 1
print(Jin.name)
print(Jin.age)
````

````python
class Dog:
  def __init__(self, name, age, family = '진돗개'):
    print(self, 'init 호출')
    self.name = name
    self.age = age
    self.family = family
````

````python
Jin = Dog('진순이', 1)
print(Jin.name)
print(Jin.age)
print(Jin.family)

Tomi = Dog('토미', 1, '시고르자브르종')
print(Tomi.name)
print(Tomi.age)
print(Tomi.family)
````

### 1-4. 메소드 정의하기
- 해당 클래스의 객체에서만 호출가능한 함수
- 해당 객체의 속성에 대한 연산을 행함
- 객체이름.메소드명() 형태로 호출됨


````python
class Counter:
  def __init__(self):
    self.number = 0
  def increment(self):
    self.number += 1
  def current_value(self):
    return self.number
  def reset(self):
    self.number = 0
````

````python
SHbank = Counter()
print(SHbank.number)

SHbank.increment()
SHbank.increment()
SHbank.increment()
SHbank.increment()

print(SHbank.number)
print(SHbank.current_value())
print('대기인원 : %d' % SHbank.current_value())
````

### 1-5. 메소드 타입
- instance method : 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
- class method :  class로 호출(함수 선언 위에 @staticmethod라고 표기)

````python
class Math:
  def add(self, x, y):
    return x + y
  def multiply(self,x, y):
    return x * y
````

````python
math = Math()
````

````python
result1 = math.add(10, 3)
print(result1)

result2 = math.multiply(3, 6)
print(result2)
````

````python
class Math:
  @staticmethod
  def add(x, y):
    return x + y
  @staticmethod
  def multiply(x, y):
    return x * y
````

````python
result1 = Math.add(10, 5)
print(result1)

result2 = Math.add(3, 6)
print(result2)
````
