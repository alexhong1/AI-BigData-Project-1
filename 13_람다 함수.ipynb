# 1. 람다 함수

- 단일문으로 표현되는 익명함수
- 코드상에서 한번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용
- 함수 또한 클래스를 통해 생성된 객체
- return 형

```python
def square(x):
  return x ** 3
```

```python
print(square(5))
```

```python
lambda x: x ** 3
```

```python
square = lambda x: x** 3
```

```python
print(square(5))
```

```python
(lambda x: x** 3)(5)
```

```python
def myData(str):
  return len(str.strip())
```

```python
print(myData('            hogeun       '))
```

```python
data = ['watermelon    ', '      apple', '   banana   ', '      melon123']
```

```python
# sorted() : 정렬하는 함수, key속성은 어떤 방식으로 정렬을 할 것인지를 저장
print(sorted(data))
```

```python
print(sorted(data, key=myData))
```

```python
# 람다 함수 사용
print(sorted(data, key=lambda str: len(str.strip())))
```

# 2. filter 함수

- 람다가 유용하게 사용되는 대표적인 함수
- 특정 조건을 만족하는 요소만 남기고 필터링 시킴

```python
num = [1, 2, 3, 4, 5, 6, 9, 10, 14, 15, 18, 20]

num1 = list(filter(lambda n:n%2 == 0, num))
print(num1)
```

# 3. map 함수

- 각 원소를 주어진 수식에 따라 변형하여 새로운 값을 반환

```python
num = [1, 2, 3, 4, 5, 6, 9, 10, 14, 15, 18, 20]

per = list(map(lambda n: n * 0.02, num))
print(per)
```
