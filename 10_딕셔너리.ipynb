# 1. 딕셔너리(Dictionary)
- 대응관계를 나타내는 자료형으로 key와 value라는 것을 한 쌍으로 갖는 형태
- 하나의 딕셔너리의 key는 중복될 수 없음
- 하나의 딕셔너리의 value는 중복될 수 있음 

### 1-1. 딕셔너리 만들기

```python
dic1 = {}
print(type(dic1))
```

```python
dic2 = {1:'류호근', 2:'홍길동', 3:'강감찬', 4:'하모'}
print(dic2)
```

### 1-2. key를 통해 value 찾기

```python
dic2 = {1:'류호근', 2:'홍길동', 3:'강감찬', 4:'하모'}
print(dic2[1])
print(dic2[3])
```

```python
dic3 = {'no':1, 'userid':'apple', 'name':'류호근', 3:'010-0000-0000'} 
print(dic3['no'])
print(dic3[3])
```

### 1-3. 데이터 추가 및 삭제

```python
dic4 = {1:'apple'}
print(dic4)

dic4[100] = 'orange'
print (dic4)

dic4[50] = 'banana'
print(dic4)

# 수정
dic4[1] = 'lemon'
print(dic4)

# 삭제
del dic4[100]
print(dic4)
```

### 1-4. 딕셔너리 함수

```python
dic3 = {'no':1, 'userid':'apple', 'name':'류호근', 3:'010-0000-0000'} 

# keys() : key 리스트를 반환
print(dic3.keys())

# values() : value 리스트를 반환
print(dic3.values())

# items() : key와 value를 한 쌍으로 묶는 튜플을 반환
print(dic3.items())

print(dic3['userid']) # apple
# print(dic3['age']) # KeyError : 'age'

# get() : key를 이용해서 value를 반환
print()
print(dic3.get('userid')) # apple
print(dic3.get('age')) # None
print(dic3.get('age', '나이를 알 수 없음, 데이터가 없음'))

# in : key가 딕셔너리 안에 있는지 확인
print()
print('name' in dic3) # True
print('age' in dic3) # False
```

### 1-5. 딕셔너리와 for

```python
dic3 = {'no':1, 'userid':'apple', 'name':'류호근', 3:'010-0000-0000'} 

for i in dic3:
  print(i, end=' ') # 키만 복사

print()

for i in dic3.keys():
  print(i, end=' ') # 키만 복사

print()

for i in dic3.values():
  print(i, end=' ') # 값만 복사

print()
for i in dic3:
  print(dic3[i], end=' ')

print()

for i in dic3.keys():
  print(dic3.get(i), end=' ')
```

### 문제
dic3에서 모든 key 중 value가 'apple'이 있는지 확인하는 프로그램을 작성해보자!

```python
for i in dic3:
  if dic3[i] == 'apple':
    print('OK! 있네요!')
  else:
    print('NO! 없네요!')
```

### 문제
아래와 같은 딕셔너리 자료에서 주민등록번호를 추출하여 성별을 알아내는 프로그램을 작성해보자. 단, ssn의 7번째 자리의 성별은 1,3은 '남자', 2,4는 '여자라고 함'

> user = {"name":"류호근", "hp":"010-1111-0000", "ssn":"0208293148567"}

```python
user = {"name": "류호근", "hp": "010-1111-0000", "ssn": "0208293148567"}
ssn = user.get('ssn')
# print(ssn)

if 'ssn' in user:
  if ssn[6] == '1' or ssn[6] == '3':
    print('남자')
  elif ssn[6] == '2' or ssn[6] == '4':
    print('여자')
  else:
    print('아직... 안태어났을텐데?')
else:
  print('ssn이 없는데용!')
```
