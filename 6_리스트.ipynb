```python
li1 = [1, 3, 5, 7, 9]
print(li1)

# list() 함수 사용
li2 = list([1, 3, 5, 7, 9])
print(li2)

li3 = ['류호근', '최재환', '김강유', '지정현', '박상현']
print(li3)

li4 = [1, 1.5, '포근이', True]
print(li4)

li5 = [1, 2, '포그이', ['호뜩이', '류리멘탈']]
print(li5)
```

# 4. 리스트 연산자

```python
li7 = [10, 20, 30]
li8 = [40, 50, 60]

print(li7 + li8)
print(li8 + li7)

# print(li7 - li8) # TypeError: unsupported operand type(s) for -: 'list' and 'list'
# print(li7 * li8) # TypeError: can't multiply sequence by non-int of type 'list'
# print(li7 / li8) # TypeError: unsupported operand type(s) for /: 'list' and 'list'

print(li7 * 3)

li9 = [10, '하핫', 30]
print(li9)
print(li8 + li9)
print(li9[0] + li9[2])
# print(li9[1] + li9[2]) # TypeError: can only concatenate str (not "int") to str
print(li9[1] * 3)
```

# 5. 리스트의 수정/삭제

```python
li10 = [10, 20, 30]
print(li10)
li10[1] = 100
print(li10)

# li10[1] # 코드셀 마지막에 넣으면 print 함수 없이도 출력이 되네?
```

```python
li10 = [10, 20, 30]

# 슬라이싱을 사용해서 값을 수정할 때
li10[1:2] = [55, 66, 77] # 슬라이싱을 이용하여 리스트 타입으로 수정한 경우 리스트에 데이터가 들어간다.
print(li10) 

# 인덱싱을 사용해서 값을 수정할 때
li10 = [10, 20, 30]
li10[1] = [55, 66, 77] #  인덱싱을 이용하여 리스트 데이터를 추가한 경우 리스트 안에 리스트가 들어간다.
print(li10)

print(li10[0:3])
# li10[1:3] = [] # 빈 리스트를 저장 -> 삭제
# print(li10)

li10[2] = [] 
print(li10)

del li10[0:3]
print(li10)
```

# 6. 리스트 함수

```python
li11 = [10, 20, 30]
print(li11)

# li11[3] = 40 # IndexError

# append() : 리스트 마지막 자리에 데이터를 추가
li11.append(40)
print(li11)

li11.append([50, 60])
print(li11)

# li11.append([70,80], 90) # TypeError
# print(li11)
```

```python
li12 = [10, 20, 30, 40, 50]
print(li12)

# pop() : 리스트 마지막에 저장된 데이터를 반환하고 해당 데이터는 삭제
# 매개변수로 인덱스를 전달하면 해당 인덱스의 데이터가 삭제.
li12.pop(1)
print(li12)
temp = li12.pop()
print(li12)
print(temp)
```

```python
li13 = [10, 20, 30, 40, 50, 20, 40]

# remove() : 매개변수로 전달된 데이터를 제거. 같은 데이터가 여러개 있는 경우 처음 출현하는 데이터를 삭제
li13.remove(20)
print(li13)

li13.remove(20)
print(li13)

# li13.remove(20) # ValueError
```

```python
li14 = [10, 20, 30]

# insert() : 리스트의 특정 위치(인덱스)에 데이터를 삽입
li14.insert(1, 100)
print(li14)
```

```python
li15 = [10, 20, 30]

# index() : 리스트에 저장된 데이터가 있을 경우 위치를 인덱스로 반환(없으면 에러)
print(li15.index(30))
# print(li15.index(100)) # ValueError: 100 is not in list
```

```python
li16 = [45, 88, 66, 4, 23, 47]

# reverse() : 리스트에 저장된 데이터를 반대 순서로 반환
li16.reverse()
print(li16)
```

```python
li17 = [10, 40, 30, 100, 90, 50]

# sort() : 리스트에 저장된 데이터를 오름차순, 내림차순으로 정렬
li17.sort() # 오름차순
print(li17)

# sort(reverse = True) # 내림차순
li17.sort(reverse = True)
print(li17)

li18 = ['Apple', 'apple', 'banana', 'Banana', 'orange', 'grape']
li18.sort(reverse = True)
print(li18)

li19 = ['류호근', '강동원', '장동건', '김도진', '디카프리오']
li19.sort()
print(li19)
```

```python
li18 = ['Apple', 'apple', 'banana', 'Banana', 'orange', 'grape']
# result = sorted(li18)
result = sorted(li18, reverse=True)
print(result)
print(li18)
```

```python
li19 = [10, 20, 30, 50, 20, 40, 30, 20]

# count() : 리스트에 포함된 특정 데이터의 갯수를 반환
print(li19.count(20))
print(li19.count(15))
```

```python
li20 = [1, 2, 3]

# extend() : 리스트를 더함
li20.extend([40, 50])
print(li20)

li21 = [1, 2, 3]
# li21 = li21 + [40, 50]
li21 += [40, 50] # 복합대입연산자
print(li21)
```

```python
# len() : 요소의 갯수 반환
print(len(li21))

# 9시부터
```
