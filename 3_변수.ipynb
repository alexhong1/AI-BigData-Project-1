# 1. 변수

- 변수란 처리할 데이터를 저장시키는 기억장소

```python
name = '류호근'
age = 22

print(name)
print(age)

name = '백지호'
print(name)
```

# 2. 변수의 이름을 작성하는 방법

- 값을 나타내는 핵심단어
- 반드시 문자로 시작하고 예약어는 사용할 수 없음(예약어 : 이미 쓰임이 있는 단어)
- 영문자, 숫자, 특수문자(_)로 사용하며 대소문자를 구별

```python
num1 = 10
num2 = 3

print('%d / %d = %06.2f' % (num1, num2, num1/num2))
```

# 3. 변수의 데이터 타입

- 데이터 타입(자료형) : 자료가 저장되는 형태를 의미

```python
name = '류호근'
age = 22
weight = 55.5
isLover = False

# type() : 변수의 데이터 타입을 반환해주는 함수
print(type(name)) # <class 'str'>
print(type(age)) # <class 'int'>
print(type(weight)) # <class 'float'>
print(type(isLover)) # <class 'bool'>
```

```python
print(isLover)
```

```python
# 변수를 삭제
del isLover
```
