# 1. 모듈
- 클래스 또는 함수를 모아 놓은 파일
- 모든 파이썬 파일(.py)은 모듈
```
# 모든 모듈의 전체 함수 가져오기
import 모듈명
# 사용
모듈명.함수명()
```

````python
import util
````

````python
util.total(1, 100)
````

```
# 특정 함수만 골라서 가져오기
from 모듈명 import 함수1, 함수2 ...
# 사용
함수명()
```

````python
from util import calc
````

````python
_,_,result,_ = calc(5,8)
print(result)
````

```
# 모든 모듈의 함수 가져오기
from 모듈명 import *
# 사용
함수명()
```

````python
from util import *
````

````python
total(1, 100)
````

```
# 모듈 별명 붙이기
import 모듈명 as 별명
# 사용
별명.함수명()
```

````python
import util as ut
````

````python
ut.total(1, 100)
````

# 2. 패키지
- 모듈의 상위개념
- 모듈이 모여있는 디렉토리

### 2-1. 패키지 관리자
- 외부 모듈을 설치  또는 삭제하기 위한 프로그램
- pip 명령을 사용(파이썬 3.4 이 후 파이썬을 설치하면 pip가 함께 설치됨)
- 윈도우 명령 프롬프트 또는 터미널에서 사용
- pip-version 으로 버전을 확인할 수 있음
